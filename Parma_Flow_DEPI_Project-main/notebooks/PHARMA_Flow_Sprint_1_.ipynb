```python
# ════════════════════════════════════════════════
#  DEMAND & SALES — Cleaning
# ════════════════════════════════════════════════

demand_clean = df_demand.copy()

# Step A: Remove duplicates
demand_clean.drop_duplicates(inplace=True)

# Step B: Fix data types
demand_clean['date'] = pd.to_datetime(demand_clean['date'], errors='coerce')

numeric_cols = ['actual_demand_units', 'dispensed_units', 'returned_units',
                'net_sales_value_egp', 'unit_price_egp',
                'flu_index', 'gastroenteritis_index', 'allergy_index']
for col in numeric_cols:
    demand_clean[col] = pd.to_numeric(demand_clean[col], errors='coerce')


# Step C: Handle missing values
for col in demand_clean.columns:
    if demand_clean[col].isnull().sum() > 0:
        if pd.api.types.is_string_dtype(demand_clean[col]) or demand_clean[col].dtype == 'object':
            demand_clean[col] = demand_clean[col].fillna(demand_clean[col].mode()[0])
        else:
            demand_clean[col] = demand_clean[col].fillna(demand_clean[col].median())


print(f"Total missing: {demand_clean.isnull().sum().sum()}")

# Step D: Strip whitespace from text columns
str_cols = demand_clean.select_dtypes(include=['object', 'string']).columns
demand_clean[str_cols] = demand_clean[str_cols].apply(lambda x: x.str.strip())



print(f"\n Shape:{demand_clean.shape[0]:,} rows * {demand_clean.shape[1]} columns")
print(f" Duplicates: {demand_clean.duplicated().sum()} rows")
print(f" Missing Values: {demand_clean.isnull().sum().sum()} total")
print(f"\n Data Types (selected):")
print(demand_clean[['date', 'actual_demand_units', 'net_sales_value_egp']].dtypes)

```

# **Egypt Pharmaceutical Demand & Sales Dataset**

## Overview:
This dataset captures pharmaceutical sales activity across Egypt throughout the full year of 2025 — from January to December, spanning 27 governorates and 64 facilities across three outlet types. Each row represents a single drug SKU sold at a single facility on a single day, making the data highly granular and well-suited for demand forecasting and regional analysis.

> 79=Drug SKUs

> 16,080=Total records

> 22=Therapeutic categories

> EGP 555 = Avg. daily net sales

## 1.   Dataset structure

The 25 columns fall naturally into five groups. The first tells us when and where: date, season, and governorate. The second identifies who made the sale: the facility, its name, and its type. The third describes what was sold: generic name, brand name, therapeutic category, dosage form, strength, pack size, and manufacturer. The fourth captures external context: disease indices, seasonal flags, promotions, and holidays. The fifth is the outcome of all the above: demand, dispensed units, and revenue.

*Time & location

*Facility identity

*Product details

*External factors

*Sales & demand



## 2.   **Target variables — what we're trying to understand**

There are four columns measuring demand and sales, and the relationships between them matter before any modelling begins. actual_demand_units is the recorded or expected demand — it contains 240 missing values, likely due to inconsistent logging at certain facilities. dispensed_units is what was actually given out, with a near-perfect correlation to demand (≈1.00), suggesting the two are almost always identical. returned_units is negligible — less than 0.1% of sales. And net_sales_value_egp is simply (dispensed − returned) × unit price.

**Technical note:** Having both actual_demand and dispensed_units as separate columns implies the dataset distinguishes between what was requested and what was fulfilled — a distinction that becomes meaningful in scenarios involving stockouts or dispensing restrictions.

## 3.   **External factors — the smartest part of the dataset**
This is the most analytically interesting section. Six columns encode external context that most pharma datasets simply don't include:

*flu_index and gastroenteritis_index carry a strong negative correlation (−0.82). This makes biological sense — flu season peaks in winter while GI illness tends to spike in summer. Using both together in a regression model without treatment would introduce multicollinearity.

*ramadan_flag is encoded separately from public holidays — a deliberate and smart choice, since Ramadan affects dispensing patterns differently, particularly for chronic medications like diabetes and hypertension treatments.

*promotion_flag appears in only ~10.7% of records, meaning promotions are unevenly distributed. It will be worth investigating whether promotions shift timing of purchases or actually generate net new demand.

## 4. **Missing data — and what it's telling us**

Three columns contain missing values, and the pattern is not random. brand_name is missing in 160 rows — these are almost certainly generic products with no commercial brand. unit_price_egp is missing in the exact same 160 rows, confirming the two are linked to the same records. actual_demand_units is missing in 240 rows — a slightly different set, which points to specific facilities that log dispensing but not formal demand requests.

## 5. **Key correlations — the numbers worth knowing**

Three correlation pairs from the numeric matrix stand out:

*unit_price ↔ net_sales: +0.47 — Higher-priced drugs generate more revenue even at lower volumes. Pricing strategy is visibly encoded in the data.

*unit_price ↔ actual_demand: −0.31 — More expensive drugs see lower demand, as expected. But −0.31 is moderate, not overwhelming — price isn't the only driver.

*actual_demand ↔ net_sales: +0.48 — Higher volume means higher revenue. The intuition holds, and the magnitude is consistent with the price effect above.

# **Egypt Pharmaceutical Inventory & Stock Dataset**

## Overview:
> Jan – Nov 2025

> 14,060 records · 28 columns

This dataset is a series of inventory snapshots taken across 8 regional warehouses in Egypt over 47 distinct dates. Each row represents the stock position of one drug SKU at one warehouse on one date — capturing not just how much is on the shelf, but how long it will last, how risky the situation is, and where in the warehouse it physically lives.


* Total records=14060

* Warehouses · 8 suppliers

* SKUs tracked = 79

* Avg. stock cove =27.6 days




## 1.  **Dataset structure — what each column group does**

The 28 columns fall into five logical groups. The first establishes when and where the snapshot was taken: date, warehouse ID, name, and governorate. The second identifies the supplier behind that SKU. The third describes the drug itself: generic name, brand, therapeutic category, dosage form, strength, manufacturer, cold-chain flag, and storage temperature. The fourth is the inventory logic layer — the calculated thresholds that drive replenishment decisions. The fifth is the live stock position: what's physically on the shelf right now, how long it will last, and how risky the situation is.

* Snapshot identity
* Supplier
* Drug details
* Inventory logic
* Stock position
* Risk output




## 2. **The inventory logic columns — the core of supply chain thinking**

avg_daily_demand_30d is the rolling 30-day demand rate and serves as the baseline for every other calculated field. safety_stock_units is the buffer held against variability — it correlates strongly with demand (r = 0.84), meaning high-demand drugs carry proportionally more buffer. reorder_point_units is the trigger threshold for placing a new order, and correlates almost perfectly with safety stock (r = 1.00) — they are effectively derived from the same formula. lead_time_days averages 9.2 days, ranging from 3 to 23, reflecting real differences in supplier reliability and Egypt's regional geography.

**Formula :**

* implied by the data: reorder_point ≈ (avg_daily_demand × lead_time) + safety_stock.
* The near-perfect r = 1.00 between safety_stock and reorder_point confirms this structure holds consistently across all SKUs.



## 3. **Stock position columns — reading the shelf in numbers**

Five columns describe what is actually in the warehouse. The arithmetic is explicit:

* available_units = on_hand − reserved − damaged. stock_cover_days then divides available by daily demand — a forward estimate of how many days remain before a stockout. The average of 27.6 days sounds comfortable, but the distribution matters: some SKUs sit at 62 days of cover while others are already below their lead time. batch_count (avg 3.5) tracks how many distinct
batches are present, relevant for expiry management and FEFO rotation. damaged_units correlates with on_hand at r = 0.72, suggesting damage scales with volume rather than being concentrated in isolated incidents.

## 4. **Shortage risk — what the labels are really telling us**

The shortage_risk_level column assigns each record to low (10,156 · 72.2%), medium (3,387 · 24.1%), or high (517 · 3.7%). But computing directly from the numeric columns reveals a sharper picture:

* 3,904 records (27.8%) have available_units below their reorder_point — meaning a replenishment order should already have been triggered at the time of the snapshot.
* 106 records have stock_cover_days below lead_time_days — stock will physically run out before a new order can arrive, regardless of when it is placed.

* Worth investigating: The gap between 3,904 below-reorder records and only 517 high-risk flags suggests the label may account for orders already placed and in transit — likely captured in reserved_units. Understanding this labelling logic is essential before using shortage_risk_level as a model target.

## 5. **Cold chain — a small but sensitive minority**

861 records (6.1%) involve cold-chain products stored at 2–8°C — almost certainly vaccines, biologics, or hormonal drugs. The cold_chain_flag and storage_temp columns are perfectly colinear, so only one is needed in any model. These SKUs warrant separate analysis: their shortage consequences are more severe, their logistics are more constrained, and their lead time variability is harder to absorb.

## 6. **Data quality notes**

* 175 missing values in both avg_daily_demand_30d and on_hand_units — always the same rows, pointing to newly added SKUs with no 30-day demand history at the time of the snapshot.

# **Egypt Pharmaceutical External Factors Dataset**

## Overview
>Jan 2025 – Feb 2026

>10,840 records · 18 columns

This dataset captures the external environment that shapes pharmaceutical demand in Egypt — not what was sold or stocked, but the conditions that explain why demand moves the way it does. It is a daily, governorate-level panel covering 28 governorates across 400 dates, encoding weather, disease pressure, cultural events, and public health context into a single table designed to sit alongside the sales and inventory datasets.

* Total records =10,840

* Governorates=27

* Unique dates =400

* Columns =18



## 1. **Dataset structure — four layers of context**

The 18 columns divide into four natural groups. The first is the observation key: date, governorate, and season — uniquely identifying each row. The second is weather: temperature and humidity, which drive disease seasonality. The third is the disease and demand indices: six continuous scores capturing illness burden and healthcare utilisation. The fourth is the calendar and events

layer: seven binary flags marking school terms, public holidays, religious events, and health campaigns, plus a summary disease alert level.

* Observation key
* Weather
* Disease & demand indices
* Calendar & event flags
* Alert output


## 2. **Weather columns — the root cause of seasonal disease**

avg_temp_c ranges from 9.7°C (winter, northern governorates) to 40.3°C (summer, Upper Egypt), with a mean of 23.6°C. humidity_pct ranges from 24.6% to 83.8%, averaging 55.4%. These two columns are not just descriptive — they are the underlying drivers behind the disease indices. Temperature alone correlates at −0.92 with flu_index and +0.89 with gastroenteritis_index, meaning these three columns are nearly redundant if all are included in the same model. In practice, temperature is the more interpretable root cause; the indices are derived consequences.

Multicollinearity warning: avg_temp_c, flu_index, and gastroenteritis_index form a tight triangle of correlations (−0.92, +0.89, −0.81). Using all three together in a linear model without treatment will inflate variance estimates. Consider using temperature as the primary feature and dropping or orthogonalising the indices.


## 3. **Disease & demand indices — what each one measures**

Six continuous scores, each ranging roughly 0–100, each capturing a different dimension of healthcare pressure:

* flu_index — respiratory illness burden. Peaks in winter, strongly anti-correlated with temperature (r = −0.92). Mean 43.2, max 96.3.
gastroenteritis_index — GI illness burden. Peaks in summer heat. Mean 41.5, max 100. Negative correlation with flu_index (r = −0.81) confirms the seasonal inversion.
* allergy_index — airborne allergen and respiratory allergy pressure. Largely independent of the other indices (max r = 0.09), making it a genuinely additive feature. Mean 31.4, peaks in spring.
* chronic_rx_refill_index — demand pressure for chronic medications (diabetes, hypertension, cardiology). Narrow range (37.8–76.9), low variance (std 4.8), and near-zero correlations with all other indices — it captures a stable, year-round demand signal independent of seasons.
* tourism_pressure_index — population density uplift from tourism. Correlated with temperature (r = +0.43) and anti-correlated with flu (r = −0.48), reflecting that tourist peaks coincide with warm months. Mean 12.0, max 42.9 — relevant for Red Sea and South Sinai governorates specifically.
* pm25_index — air pollution / particulate matter. Notably independent from all other indices (all correlations near 0.00). Mean 29.6, max 64.9. Likely driven by local industrial and traffic patterns rather than seasonal factors.


4. **Calendar & event flags — discrete demand shock**

Seven binary (0/1) columns mark specific calendar conditions. Each represents a different type of demand shift:
* school_term_flag — active 69.5% of days. School terms correlate with higher paediatric infection rates and antibiotic demand.
* public_holiday_flag — active 30.3% of days. Affects dispensing patterns through facility closures and patient behaviour changes.
* ramadan_flag — active 7.2% of days (~26 days). Changes meal timing affects insulin and antacid demand significantly.
* eid_al_fitr_flag and eid_al_adha_flag — active ~1% each (108 and 136 rows). Short but sharp demand spikes, particularly for analgesics and wound care around Eid al-Adha.
* healthcare_campaign_flag — active 1.1% of days (120 rows). Marks national vaccination or screening drives, which can temporarily suppress or redirect demand at specific facilities.

* Modelling note: The Eid flags and healthcare_campaign_flag are active in fewer than 1.5% of rows each. They carry strong signal when they occur, but their low frequency means tree-based models may deprioritise them unless boosted or engineered into interaction features (e.g. ramadan × chronic_rx_refill).


5. **Disease alert level — the summary output**

disease_alert_level classifies each row as low (3,441 · 31.7%), medium (4,363 · 40.2%), or high (3,036 · 28.0%). Notably, the distribution here is far more balanced than the shortage_risk_level in the inventory dataset — high alert represents nearly a third of all rows, reflecting how frequently Egypt's governorates face elevated disease burden. This column is likely derived from a combination of flu_index, gastroenteritis_index, and possibly pm25_index, and could serve as a demand-pressure label in downstream modelling.



6. **Data quality & linkage notes**


* 72 missing values in avg_temp_c, humidity_pct, and flu_index — always the same rows, suggesting a weather station outage or reporting gap for specific governorate-date combinations rather than a systematic issue.
* Date range extends to February 2026 (400 unique dates), giving this dataset a longer horizon than the sales dataset (which ends December 2025) and the inventory dataset (which ends November 2025). The extra months could be useful for forward-looking demand scenarios.
* Join key to other datasets: this table links to the sales dataset on (date, governorate) and to the inventory dataset via (date, governorate) through the warehouse-governorate mapping. All three datasets share the same 28-governorate geography.


# **Egypt Pharmaceutical Expiry & Batch Tracking Dataset**

## Overview
As of 31 Dec 2025  
12,550 records · 24 column

This dataset tracks every individual batch of medication held across Egypt's 8 regional warehouses — from the day it arrived to its expiry date and current condition. Where the inventory dataset shows stock totals, this one goes one level deeper: each row is a distinct batch with its own receipt date, expiry date, quantities in each condition state, and a calculated financial value. It is the operational backbone of FEFO (First Expired, First Out) rotation and waste management.

* Batches tracked =12,550

* SKUs · 8 warehouses=79

* Value at risk = EGP 12.1M

* Expired with stock remaining = 734



1. **Dataset structure — four column groups**

The 24 columns organise into four layers. The first is the batch identity: a unique batch ID, the warehouse it lives in, and the drug it belongs to (SKU, name, brand, manufacturer, dosage form, strength, cold chain, and storage temp). The second is the batch timeline: receipt date, expiry date, and days_to_expiry calculated as of 31 December 2025 — negative values mean the batch has already expired. The third is the quantity breakdown: initial quantity split into remaining, damaged, quarantine, and wastage. The fourth is the output layer: batch status, expiry risk level, FEFO rank hint, and estimated financial value.
* Batch identity
* Timeline
* Quantity breakdown
* Status & risk output



2. **Batch timeline — reading the dates**

receipt_date records when the batch arrived at the warehouse (550 unique dates). expiry_date records the manufacturer's expiry (1,223 unique dates, covering a wide range of shelf lives). days_to_expiry_as_of_2025_12_31 is the key derived column: positive values = days of shelf life remaining at year-end, negative values = already expired. The range spans from −183 days (expired over 6 months ago) to +1,096 days (nearly 3 years of shelf life remaining), with a mean of 392 days. This single column drives both the batch_status classification and the expiry_risk_level label.

Calculation reference: days_to_expiry = expiry_date − 2025-12-31. A batch with days_to_expiry = −182 expired on approximately 2025-07-02. The 125 rows missing expiry_date also have missing remaining_qty, suggesting these are administrative holds or batches with incomplete intake records.


3. **Quantity breakdown — where did the units go?**

Every batch starts with an initial_batch_qty (avg 201, max 1,011). Units then flow into four states. remaining_qty is what is still physically on the shelf and usable (avg 98, strongly correlated with initial qty at r = 0.70). damaged_qty covers physically compromised units (avg 1.5, total 18,525 across the dataset). quarantine_qty covers units held pending investigation or quality release (avg 1.1, total 13,249). wastage_qty covers units disposed of due to expiry or quality failure (avg 3.0, total 38,000 — the largest loss category). Average batch utilisation across the dataset is 50.8%, meaning roughly half of each batch's initial quantity has been consumed.

Implicit formula: consumed = initial − remaining − damaged − wastage. Quantities do not always sum perfectly due to quarantine overlaps and timing of records, but the logic holds across the majority of batches.


4. **Batch status & expiry risk — the most operationally critical columns**

batch_status classifies every batch into one of four states. The distribution tells the overall picture of warehouse health:
* usable — 10,765 batches (85.8%): within shelf life, cleared for dispensing.
near_expiry — 950 batches (7.6%): approaching expiry, likely within 90 days. Priority for FEFO rotation and redistribution.
* expired — 742 batches (5.9%): past expiry date as of 31 Dec 2025. Of these, 734 still have remaining_qty > 0 — meaning nearly every expired batch still has stock on the shelf that should have been dispensed or disposed of.
* quarantine — 93 batches (0.7%): held for quality investigation, not available for dispensing.

* Critical finding: 734 expired batches still carry remaining stock, and the dataset records EGP 12.1 million in combined value across near_expiry and expired batches. This represents direct financial loss if not redistributed, returned to supplier, or formally written off. Total wastage_qty across the full dataset is 38,000 units.



5. **FEFO rank hint — the dispensing priority guide**

fefo_rank_hint assigns each batch a priority rank from 1 (dispense first — nearest to expiry) to 7 (dispense last — most shelf life remaining). It is an integer, not a continuous score, and its distribution is bell-shaped: most batches cluster at ranks 2–4, with very few at rank 7 (74 batches). The correlation with days_to_expiry is effectively zero (r = 0.01), which is expected — rank is relative within a SKU-warehouse group, not absolute across all batches. This column is not useful as a standalone feature but is essential for operational FEFO compliance checks and for validating whether dispensing order is being respected.


6. **Estimated batch value — financial exposure in one column**

estimated_batch_value_egp translates the physical stock position into financial terms. It ranges from EGP 71.7 (a small low-cost batch) to EGP 167,790 (a large high-cost drug, likely a biologic or specialty medication). The dataset's total batch value is EGP 88.1 million, of which EGP 12.1 million — 13.7% — is in batches that are already expired or near expiry. This column correlates most strongly with remaining_qty (r = 0.55) rather than initial_qty (r = 0.38), reflecting that value is calculated on what remains, not what was received.



7. **Data quality notes**

* 125 rows missing expiry_date and remaining_qty simultaneously — the same records, pointing to batches with incomplete intake processing. These should be excluded from expiry analysis or imputed cautiously.
* 50 duplicate batch_ids — 12,550 rows but only 12,500 unique batch IDs. Duplicates likely represent either data entry errors or batches that were split across two warehouse locations. Needs investigation before using batch_id as a join key.
* cold_chain_flag and storage_temp are colinear (same as the inventory dataset) — 815 cold-chain batches at 2–8°C. These deserve separate treatment in any expiry analysis since cold-chain failures accelerate degradation.

* Join key to other datasets: this table links to the inventory dataset on (warehouse_id, sku_id) — batch_count in the inventory dataset should match the number of active batch rows here per warehouse-SKU combination. It links to the sales dataset on (sku_id, warehouse → governorate) to connect dispensing velocity with expiry risk.

## Demand Forecasting Models

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# ── Demand Forecasting Models (Regression) ──────────────────────────
models_demand = {
    "Random Forest":         RandomForestRegressor(n_estimators=200, max_depth=12, n_jobs=-1, random_state=42),
    "Gradient Boosting":     GradientBoostingRegressor(n_estimators=200, max_depth=6, learning_rate=0.05, random_state=42),
    "Ridge Regression":      Ridge(alpha=1.0),
}

demand_results = {}
for name, model in models_demand.items():
    model.fit(X_train_demand, y_train_demand)
    y_pred = model.predict(x_test_demand)
    mae  = mean_absolute_error(y_test_demand, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test_demand, y_pred))
    r2   = r2_score(y_test_demand, y_pred)
    demand_results[name] = {"MAE": round(mae,2), "RMSE": round(rmse,2), "R2": round(r2,4)}
    print(f"{name:25s}  MAE={mae:.2f}  RMSE={rmse:.2f}  R²={r2:.4f}")

import pandas as pd
demand_results_df = pd.DataFrame(demand_results).T
print("\n--- Demand Model Comparison ---")
print(demand_results_df.sort_values("R2", ascending=False))
```

```python
# ============================================================
# Step 8: Final Output
# Build the unified risk output per SKU and export to JSON/CSV
# ============================================================

import json
from datetime import datetime

# ── 1. Shortage Risk: from inventory_clean ──────────────────
shortage_df = inventory_clean[[
    'sku_id', 'generic_name', 'available_units',
    'avg_daily_demand_30d', 'stock_cover_days',
    'reorder_point_units', 'shortage_risk_level'
]].copy()

# Forecast demand = avg daily demand × 30 days
shortage_df['forecast_demand'] = (
    shortage_df['avg_daily_demand_30d'] * 30
).round(0)

shortage_df.rename(columns={
    'available_units':      'stock',
    'shortage_risk_level':  'shortage_risk'
}, inplace=True)

# ── 2. Expiry Risk: from expiry_clean ───────────────────────
expiry_agg = expiry_clean.groupby('sku_id').agg(
    expiry_days   = ('days_to_expiry_as_of_2025_12_31', 'min'),   # worst batch
    expiry_risk   = ('expiry_risk_level', lambda x: x.value_counts().idxmax()),
    remaining_qty = ('remaining_qty', 'sum')
).reset_index()

# ── 3. Merge both on sku_id ──────────────────────────────────
final_df = shortage_df.merge(expiry_agg, on='sku_id', how='left')

# ── 4. Unified Risk Label ────────────────────────────────────
def unified_risk(row):
    if row['shortage_risk'] == 'high' or row.get('expiry_risk') == 'high':
        return 'high'
    elif row['shortage_risk'] == 'medium' or row.get('expiry_risk') == 'medium':
        return 'medium'
    else:
        return 'low'

final_df['overall_risk_level'] = final_df.apply(unified_risk, axis=1)

# Determine dominant risk type
def risk_type(row):
    if row['shortage_risk'] == 'high':
        return 'shortage'
    elif str(row.get('expiry_risk', '')) == 'high':
        return 'expiry'
    elif row['shortage_risk'] in ['medium', 'high']:
        return 'shortage'
    else:
        return 'none'

final_df['risk_type'] = final_df.apply(risk_type, axis=1)

# ── 5. Select & clean final columns ─────────────────────────
output_df = final_df[[
    'sku_id', 'generic_name', 'forecast_demand', 'stock',
    'stock_cover_days', 'expiry_days', 'remaining_qty',
    'shortage_risk', 'expiry_risk', 'risk_type', 'overall_risk_level'
]].drop_duplicates(subset='sku_id').reset_index(drop=True)

print(f"✅ Total SKUs in output: {len(output_df)}")
print(f"\nRisk Distribution:\n{output_df['overall_risk_level'].value_counts()}")
print(f"\nRisk Types:\n{output_df['risk_type'].value_counts()}")
print(f"\nSample output:")
output_df.head(3)
```
